```python
df1= pd.read_csv('C:/Users/wojci/Documents/Studia/4 rok/Complex systems/Raport 1/generated_data_files/clusters_dist/L=100/Dist_p0.1L100T10000.txt', 
                   sep=" ", header=None, names=['frequency','size'])
df2= pd.read_csv('C:/Users/wojci/Documents/Studia/4 rok/Complex systems/Raport 1/generated_data_files/clusters_dist/L=100/Dist_p0.2L100T10000.txt', 
                   sep=" ", header=None, names=['frequency','size'])
df3= pd.read_csv('C:/Users/wojci/Documents/Studia/4 rok/Complex systems/Raport 1/generated_data_files/clusters_dist/L=100/Dist_p0.3L100T10000.txt', 
                   sep=" ", header=None, names=['frequency','size'])
df4= pd.read_csv('C:/Users/wojci/Documents/Studia/4 rok/Complex systems/Raport 1/generated_data_files/clusters_dist/L=100/Dist_p0.4L100T10000.txt', 
                   sep=" ", header=None, names=['frequency','size'])
df5= pd.read_csv('C:/Users/wojci/Documents/Studia/4 rok/Complex systems/Raport 1/generated_data_files/clusters_dist/L=100/Dist_p0.5L100T10000.txt', 
                   sep=" ", header=None, names=['frequency','size'])

plt.scatter(df1['size'],np.log(df1['frequency'].div(10000)),marker="o")
plt.scatter(df2['size'],np.log(df2['frequency'].div(10000)),marker="*")
plt.scatter(df3['size'],np.log(df3['frequency'].div(10000)),marker="^")
plt.scatter(df4['size'],np.log(df4['frequency'].div(10000)),marker=".")
plt.scatter(df5['size'],np.log(df5['frequency'].div(10000)),marker="+")


plt.legend(["p=0.1","p=0.2","p=0.3","p=0.4","p=0.5"])
plt.title('Cluster dsitribution, T=10000,L=100')
plt.xlabel('size')
plt.ylabel("log n")

a=plt.gcf()
a.savefig("C:/Users/wojci/Documents/Studia/4 rok/Complex systems/Raport 1/cluster_distribution100_low_p.eps")
plt.show()
```
